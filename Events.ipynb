```python
import pandas as pd
import numpy as np
```

```python
#type file names here
ua_file_name = 'Acura_UA_Events.csv'
g4_file_name = 'Acura_GA4_Events.csv'
```

```python
def g4_events_to_ua(event_name):
    if event_name in ["video_start","video_pause","video_progress","video_complete"]:
        return 'video'
    if event_name in ["scroll"]:
        return 'scroll depth'
    if event_name in ["nav_link_click"]:
        return 'navigation'
    if event_name in ["allow_geo_features","deny_geo_features"]:
        return 'geodetection'
    if event_name in ["form_error","form_visible","form_completion"]:
        return 'forms'
    if event_name in ["404_error"]:
        return '404 error'
    if event_name in ["chat_with_a_specialist_icon_click", "start_chat_request", "cancel_chat_request", "chat_started", "chat_ended", "view_chat_cta_modal", "chat_with_a_specialist_modal_link_click", "close_chat_with_a_specialist_modal"]:
        return 'chat'
    if event_name in ["social_exit_link"]:
        return 'social'
    if event_name in ["file_download"]:
        return 'file download'
    if event_name in ["google_survey_loaded"]:
        return 'google survey'
    if event_name in ["hero_banner_click"]:
        return 'hero banner'
    if event_name in ["promo_tile_click", "promo_tile_impression", "secondary_banner_click"]:
        return 'promotions'
    if event_name in ["build_and_price_cta_click", "view_product_details", "book_a_test_drive_cta_click", "book_an_appointment_cta_click", "view_showroom_cta_click","view_product_details", "filter_by_vehicle_type"]:
        return 'vehicle list'
    if event_name in ["scroll_1500px", "nav_link_click","carousel_slide_change","feature_slide_chage","brochure_download","filter_by_trim","switch_view","build_and_price_cta_click","view_product_details","cta_link_click","locate_a_dealer","book_a_test_drive_cta_click","change_exterior_color","expand_accordion_content","zoom_image","show_exterior_gallery","show_interior_gallery","feature_drawer_expand","feature_drawer_rotate","select_trim","view_showroom_cta_click","model interactions_cta_click","model interactions_tab_click"]:
        return 'model interactions'
    if event_name in ["build_and_price_cta_click","view_product_details","filter_by_body_style","filter_by_model","filter_by_trim","change_list_view_type"]:
        return 'virtual showroom'
    if event_name in ["filter_by_distance","filter_by_year","filter_by_vehicle_type","view_all_model_offers","view_product_details_click","build_and_price_cta_click","carousel_pause","carousel_resume","carousel_click","carousel_slide_change","contact_dealer"]:
        return 'offer interactions'
    if event_name in ["bap_filter_by_vehicle_type","bap_select_model", "bap_filter_by_transmission_type", "bap_view_trim_features", "bap_select_trim", "bap_select_color", "bap_filter_by_accessory_type","bap_view_acessory_details","bap_add_accessory","bap_remove_accessory", "bap_select_warranty","bap_view_warranty_details","bap_add_protection","bap_remove_protection","bap_view_protection_details","bap_complete_build","bap_edit_build","bap_brochure_download","apply_for_credit_cta_click","contact_dealer","view_my_selections_cta_click","reserve_now_cta_click","place_deposit_cta_click","bap_view_showroom_cta_click","bap_view_model_features"]:
        return 'bap'
    if event_name in ["save_build_start","save_build_complete"]:
        return 'save build'
    if event_name in ["search_dealer","view_dealer_details","get_dealer_directions","schedule_service","contact_dealer"]:
        return 'contact dealer'
    if event_name in ["oca_application_start","oca_application_progress","oca_input_error","oca_edit_details","oca_application_complete","oca_resume_application","oca_suspend_application","oca_print_application","oca_send_application_to_dealer","oca_start_credit_application_cta_click","oca_view_aplication_details","oca_application_cancel"]:
        return 'credit application'
    if event_name in ["account_add_vehicle","account_registration_start","account_registration_complete","account_login","account_logout","number_of_builds","delete_build","view_build","apply_for_credit_cta_click","add_build_to_favorites","contact_dealer"]:
        return 'account'
    if event_name in ["contact_dealer","book_an_appointment_cta_click","search_vehicle","visit_dealer_site","view_product_details"]:
        return 'cuv'
    if event_name in ["search_dealer","payment_cancel","payment_success","payment_fail"]:
        return 'payment gateway'
    if event_name in ["future_vehicles_cta_click","future_vehicles_tab_click","future vehicles_cta_click","future vehicles _cta_click"]:
        return 'future vehicles'
    if event_name in ["contact us_cta_click"]:
        return 'contact us'
    if event_name in ["parts & service_cta_click"]:
        return 'parts & service'
    if event_name in ["financial services_cta_click"]:
        return 'financial services'
    else:
        return event_name
    
```

```python
#Read data, grab only the first table, not the date table
ua_file = pd.read_csv(ua_file_name, header = 6, thousands=',')
n_values = np.logical_not(ua_file.isnull()).sum(axis=1)
cut_off = n_values[n_values == 0].index[0]
ua_file = ua_file[:cut_off -1]
ua_file['Total Events'] = ua_file['Total Events'].str.replace(',', '')
g4_file = pd.read_csv(g4_file_name, header = 6, thousands=',')
g4_file['Event name'] = g4_file['Event name'].apply(g4_events_to_ua)
g4_file.rename(columns={'Event name': 'Event Category', 'Total users': 'Total Events'}, inplace = True)
g4_file = g4_file[['Event Category', 'Total Events']]
g4_file = g4_file[1:]

#Run the code below (get rid of the hashtag) if the 'Users' or 'Sessions' columns have commas
ua_file['Total Events'] = ua_file['Total Events'].str.replace(',', '')
# g4_file['Total Events'] = g4_file['Total Events'].str.replace(',', '')

#Add data source column

ua_file['data_source'] = 'UA'
g4_file['data_source'] = 'G4'

#Append data, reset index
data = ua_file.append(g4_file).reset_index()
data.drop(columns={'index'}, inplace = True)
data[data['data_source'] == 'UA']
```

```python
piv = pd.pivot_table(data, values = 'Total Events', index = 'Event Category', columns = 'data_source', aggfunc = sum)
piv.columns = [c + ' Events' for c in piv.columns]
piv['G4 Events'] = piv['G4 Events'].astype('float')
piv['UA Events'] = piv['UA Events'].astype('float')
piv['Events Difference'] = piv['G4 Events'] - piv['UA Events']
piv['Events Variance (%)'] = (piv['G4 Events'] - piv['UA Events'])/ piv['UA Events']
piv = piv[[ 'G4 Events', 'UA Events', 'Events Difference', 'Events Variance (%)']]
piv.sort_values(by = ['G4 Events'], ascending = False, inplace = True)
piv
```

```python
#Export to Excel
piv.to_excel('Acura_Pivot_Events.xlsx')
```
